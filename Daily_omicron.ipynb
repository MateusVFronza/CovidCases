## Explore


### Analysing data about Brazil

```python
tabela_brazil = tabela[tabela.location == 'Brazil']
display(tabela_brazil)
```

```python
# print of all dates with data
for i in tabela_brazil.date.unique():
  print(i)
```

```python
# counting the cases reported at the most recent date, segregated by variant
tabela_brazil[tabela_brazil.date == '2021-12-27'].value_counts()
```

```python
grafico = px.histogram(tabela_brazil, 
                       x = 'date',
                       y = 'num_sequences',
                       color = 'variant',
                       title = "Daily cases of Covid-19 by variants<br><sup>Brazil 2021</sup>",
                       labels = {'date':"Date",'variant':'Variants'})
grafico.show()
```

```python
# export 
grafico.write_image(file = '/content/drive/MyDrive/Colab Notebooks/Dataset/Covid-variants/Output/brazilian_cases',
              format = 'png')
```

The plot shows the growth of delta cases since July/2021 and the rapid spread to the end of the year (Nov/2021). At the end of the year it also shows the beginning of Omicron in the last bar.

### Analyzing data from other countries

```python
# overview of dates
for i in tabela.date.unique():
  print(i)
```

```python
# to plot complete information, at first I'll list all non-null countries on the first day of 2021
list_country = [] # and make a list
for i in tabela[tabela.date =='2022-01-05'].location.unique():
  print(i)
  list_country.append(i)
```

```python
tabela_select = pd.DataFrame(tabela) #creating a df
tabela_select = tabela_select[tabela_select.location.isin(list_country)] #filter by countries with data at first date on 2021
display(tabela_select) # print the df
```

```python
# ploting the data
grafico = px.histogram(tabela_select, 
                       x = 'date',
                       y = 'num_sequences',
                       color = 'variant',
                       title = "Daily cases of Covid-19 by variants<br><sup>Bangladesh, Belgium, Czechia, France, Mexico, Netherlands, New Zealand, Pakistan, Poland, Sri Lanka, United Kingdom, United States</sup>",
                       labels = {'date':"Date",'num_sequences':'Cases','variant':'Variants'})
grafico.show()
```

```python
 #export
grafico.write_image(file = '/content/drive/MyDrive/Colab Notebooks/Dataset/Covid-variants/Output/world_cases',
              format = 'png')
```

The graph show waves of cases: 
*   Alfa from jan2021 to jul2021
*   Delta from jun2021 to jan2022
*   The beggining of omicron in late December 2021 and its rapid spread
